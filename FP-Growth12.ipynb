```python
import pandas as pd
import numpy as np
import time
import matplotlib.pyplot as plt
from sklearn import datasets
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
from mlxtend.preprocessing import TransactionEncoder
from apyori import apriori as asp_apr
```

```python
df = pd.read_excel('C://Users//user//Desktop//trade.xlsx')
df.head()
```

```python
market_data = [['MEMORY_EMBEDED','CPU / MPU','DISCRETE','PEMCO','LOGIC IC','LINEAR IC','OPTICAL AND SENSOR'],
 ['MEMORY_EMBEDED','CPU / MPU'],
 ['DISCRETE','PEMCO','LOGIC IC','LINEAR IC'],
 ['DISCRETE','PEMCO','LOGIC IC'],
 ['MEMORY_EMBEDED','CPU / MPU','DISCRETE','PEMCO','LOGIC IC'],
 ['LINEAR IC','OPTICAL AND SENSOR'],
 ['LOGIC IC'],
 ['DISCRETE','PEMCO'],
 ]
```

```python
from mlxtend.preprocessing import TransactionEncoder

te=TransactionEncoder()
te_ary = te.fit(market_data).transform(market_data)
df=pd.DataFrame(te_ary,columns=te.columns_)
df
```

```python
#['support','confidence','lift']

number = 'confidence'
threshold = 0.5
min_support = 0.01
```

```python
frequent_itemset=fpgrowth(df,min_support=min_support,use_colnames=True)
rules=association_rules(frequent_itemset,metric=number,min_threshold=threshold)

print('Input:'+str(list(rules['antecedents'].unique())))
print(' ')
print('==================================================')
print('==================================================')
print(' ')
print('Output:'+str(list(rules['consequents'].unique())))
```

```python
X = input('請輸入要購買的商品:')

empty_list = []
y = list(rules[rules['antecedents']==frozenset({X})]['consequents'].unique())
for i in range(len(y)):
    empty_list.append(list(y[i]))
    
print('額外推薦:'+str(empty_list))
```
